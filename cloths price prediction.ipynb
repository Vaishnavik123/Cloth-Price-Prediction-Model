```python
input_cols=train_df.columns.tolist()[0:-1]
target_col='price'
```

```python
train_inputs=train_df[input_cols]
train_target=train_df[target_col]
```

```python
valid_inputs=valid_df[input_cols]
valid_target=valid_df[target_col]
```

```python
test_inputs=test_df[input_cols]
test_target=test_df[target_col]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(categories='auto',sparse_output=False)
```

```python
con_df_1=pd.concat([train_df,valid_df],ignore_index=True)
```

```python
full_df=pd.concat([test_df,con_df_1],ignore_index=True)
```

```python
encoder.fit(full_df[input_cols])
```

```python
encoded_cols=encoder.get_feature_names_out(input_cols).tolist()
```

```python
train_inputs[encoded_cols]=encoder.transform(train_inputs[input_cols])
valid_inputs[encoded_cols]=encoder.transform(valid_inputs[input_cols])
test_inputs[encoded_cols]=encoder.transform(test_inputs[input_cols])
```

```python
X_train=train_inputs[encoded_cols]
```

```python
X_valid=valid_inputs[encoded_cols]
X_test=test_inputs[encoded_cols]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state=42,n_estimators=100)

model.fit(X_train,train_target)
```

```python
model.score(X_train,train_target)
```

```python
model.score(X_valid,valid_target)
```

```python
model.score(X_test,test_target)
```

```python
l=[]
```

```python
def predict(nodes):
    model = RandomForestRegressor(random_state=42,n_jobs=-1,n_estimators=500,max_depth=18,max_leaf_nodes=60,)
    model.fit(X_train,train_target)
    train_accu=model.score(X_train,train_target)
    valid_accu=model.score(X_valid,valid_target)
    return({'nodes':nodes,'t_accu':train_accu,'valid_accu':valid_accu})
    
    
```

```python
for i in range(3,100,3):
    l.append(predict(i))
accu_df=pd.DataFrame(l)
```

```python
accu_df.sort_values('valid_accu',ascending=False)
```
